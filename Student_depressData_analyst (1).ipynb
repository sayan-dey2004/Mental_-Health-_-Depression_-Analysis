```python
sns.boxplot(x='Depression', y='Age', data=df)
plt.show()
```

```python
from scipy.stats import ttest_ind
from scipy import stats
import random

df = df[(np.abs(stats.zscore(df["Age"])) < 3)]

Dep0 = df[df['Depression'] == 0]['Age']
Dep1 = df[df['Depression'] == 1]['Age']

Dep0 = Dep0.values.tolist()
Dep1 = Dep1.values.tolist()

Dep0 = random.sample(Dep0, 100)
Dep1 = random.sample(Dep1, 100)

t_statistic, p_value = ttest_ind(Dep0, Dep1, equal_var= False)

print('t-statistic: ', t_statistic)
print('p-value: ', p_value)

if p_value < 0.05:
    print('Reject null hypothesis: There have significant different between Age wise Depression')
else:
    print('Fail to reject null hypothesis: There have no significant different between Age wise Depression')
```

```python
from logging import critical
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
plt.figure(figsize=(6,5))
total = len(df['Depression'])
ax = sns.countplot(x= 'Gender',hue="Depression", data=df)

for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.show()

c_t = pd.crosstab(df['Gender'].sample(frac=0.002, replace = True, random_state =1),df['Depression'].sample(frac=0.002, replace = True, random_state =1), margins= False)
c_t

stat,p_value,dof,expected = chi2_contingency(c_t)
print('dof',dof)
print(expected)
print('p-value',p_value)

prob = 0.95
critical = chi2.ppf(prob,dof)
print('critical= %.3f, prob=%.3f, stat=%.3f' % (critical, prob, stat))

if p_value < 0.05:
    print('Reject null hypothesis: There is a significant association between Gender and Depression (Dependent)')
else:
    print('Independent' , 'There is no significant association between Gender and Depression')
```

# **Process the Data before make the model**

```python
for col in df.columns:
  if ((df[col].dtypes != 'object') & (col != 'Depression')):
    sns.boxplot(df[col])
    plt.xlabel(col)
    plt.ylabel('Values')
    plt.show()
```

```python
    q1 = df['Age'].quantile(0.25)
    q3 = df['Age'].quantile(0.75)
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr
    df = df[(df['Age'] >= lower) & (df['Age'] <= upper)]
```

```python
sns.boxplot(df['Age'])
plt.xlabel('Age')
plt.ylabel('Values')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for col in df.columns:
  if df[col].dtypes == 'object':
    df[col] = le.fit_transform(df[col])
```

```python
X = df.drop('Depression', axis=1)
y = df['Depression']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3,random_state=42)
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test,y_pred)*100)
```

## Summary:

### Data Analysis Key Findings

*   Independent samples t-tests were performed on eight variables to assess their relationship with Depression.
*   Significant differences in means related to Depression were found for Age (p-value = 0.0014), Academic Pressure (p-value = 2.23e-13), and Financial Stress (p-value = 4.86e-09).
*   No significant differences in means related to Depression were found for CGPA (p-value = 0.714), Study Satisfaction (p-value = 0.055), or Work/Study Hours (p-value = 0.052).
*   For Work Pressure and Job Satisfaction, the t-statistic and p-value were not available, and the conclusion drawn was that there was no significant difference in means related to Depression for these variables.

### Insights or Next Steps

*   Focus further analysis on the variables showing a significant relationship with Depression (Age, Academic Pressure, and Financial Stress) to understand the nature and extent of these relationships.
*   Investigate why t-statistics and p-values were not available for Work Pressure and Job Satisfaction and re-evaluate their relationship with Depression if possible.
